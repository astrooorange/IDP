This cell sets various parameters of the lab double pendulum, and dt the time step for numerical integration. [ALREADY COMPLETE]

```python
a = 0.185 #m
b = 0.172 #m
ma = 0.044 #kg
mb = 0.019 #kg
mp = 0.022 #kg
g = 9.81 #m/s^2
dt=0.002 #s
```

The generic double pendulum Lagrangian:
$$
\mathcal{L}=\frac{1}{2}\left(A\dot{\alpha}^2+B\dot{\beta}^2+ 2 C \dot{\alpha}\dot{\beta}\cos{(\alpha-\beta)}\right)+D \cos{\alpha}+E \cos{\beta},
$$
The cell below sets values of $A$...$E$ for use in the simulations. Set these parameters to the values you calculated for the lab double pendulum. It is easiest to work in terms of $a$, $b$ $m_a$ etc set above [FILL IN THE GAPS]

```python
A= (ma*a**2/4)+(ma*a**2/12)+mb*a**2+mp*a**2
B=(mb*b**2/4)+(mb*b**2/12)
C=mb*a*(b/2)
D=g*(mb*a+mp*a+ma*a/2)
E=g*(mb*b/2)
```

We want to numerically integrate the equation of motion for the double pendulum. To do this, the code stores the state of the double pendulum is a list of four variables, state=[$\alpha$,$\beta$,$\dot{\alpha}$,$\dot{\beta}$]. This cell defines a function accels(state), which takes a state list as its argument, and returns a list of the time derivatives of the four quantities [$\dot{\alpha}$,$\dot{\beta}$,$\ddot{\alpha}$,$\ddot{\beta}$]. The second time derivatives are calculated from Lagrange's equations for the generic double pendulum Lagrangian. You should check they are right, but you do not need to edit them. [ALREADY COMPLETE]

```python
def accels(state):
    [alpha,beta,alphadot,betadot]=state
    cab = np.cos(alpha - beta)
    sab = np.sin(alpha - beta)
    alphadoubledot=((-C)*(betadot**2*B + alphadot**2*cab*C)*sab - B*D*np.sin(alpha) + cab*C*E*np.sin(beta))/(A*B - cab**2*C**2)
    betadoubledot=(C*(alphadot**2*A + betadot**2*cab*C)*sab + cab*C*D*np.sin(alpha) - A*E*np.sin(beta))/(A*B - cab**2*C**2)
    return np.array([alphadot, betadot, alphadoubledot,betadoubledot])

#积分工具1：numerical integration of 速度、加速度
```

The rate of change, [$\dot{\alpha}$,$\dot{\beta}$,$\ddot{\alpha}$,$\ddot{\beta}$], is all we need to integrate using the forward Euler method. However, for greater accuracy, we instead use the fourth order Runge-Kutta method as discussed in lectures. This next cell defines the function RK4_grad(state) which takes a state list state=[$\alpha$,$\beta$,$\dot{\alpha}$,$\dot{\beta}$] as an argument and calculates a fourth-order Runge-Kutta approximation to the time derivatives of the state, RK4grad=[$\dot{\alpha}$,$\dot{\beta}$,$\ddot{\alpha}$,$\ddot{\beta}$] that combines the acceleration at the start, middle and end-point of a time step for a more accurate numerical integration. [ALREADY COMPLETE]

```python
def RK4_grad(state):
        '''
        runge kutta 4th order
        '''
        #4th order Runge-Kutta
        #[alpha,beta,alphadot,betadot]=state
        k1 = accels(state)
        k2 = accels(state + dt*k1/2)
        k3 = accels(state + dt * k2 / 2)
        k4 = accels(state + dt * k3)
        # compute the RK4 right-hand side
        RK4grad = 1.0 / 6.0  * (k1 + 2.0 * k2 + 2.0 * k3 + k4)
        return RK4grad
# 积分工具2：对输入量进行一次到四次求导，然后进行近似（上面方法的替代）
```

Finally, we are ready to integrate. The cell below defines the function

integrate(state0,total_time),

which integrates the double pendulum equations numerically. Integration starts at  t=0 and with initial conditions state0=[$\alpha(0), \beta(0), \dot{\alpha}(0),\dot{\beta}(0)$], and continues in time steps dt (defined at the top) until it reaches $t=$total_time.

The output is a list of three lists:

simulation_data=[t_list, alpha_list, beta_list],

where t_list is a list of simulation time points, and alpha_list and beta_list are lists of alpha and beta at each time point.

The function is already written, but read it and make sure you understand it. How would you modify it to use forward Euler instead? [ALREADY COMPLETE]

```python
def integrate(state0,total_time):
    t_list=np.array([])   #This will be the list of time points
    alpha_list=np.array([])   #This will be the list of alpha positions at each time point
    beta_list=np.array([])  #This will be the list of beta positions at each time point
    state=np.array(state0)  #Initial conditions: set the state of the double pendulum to state0
    t=0

    while t<total_time:     # This is the loop that actually steps through time in increments dt to do the integration.
        state = state + RK4_grad(state)*dt # One RK4 step to work out the state dt later
        alpha_list=np.append(alpha_list,state[0]) # store the new state in the alpha, beta and t lists for output
        beta_list=np.append(beta_list,state[1])
        t_list=np.append(t_list,t)
        t = t + dt         # Increase time

    return t_list, alpha_list, beta_list

# 根据initial conditions，使用积分工具2，获得t和角度（α、β）的list
# state：α、β、α˙(0),β˙(0)
```

```python
def plot_poincare_section(Simulation_Data):
    [t,alpha,beta]=Simulation_Data

    alpha=np.mod(alpha+np.pi/2,np.pi)-np.pi/2 # map angles to between +/- pi if the rod flips
    beta=np.mod(beta+np.pi/2,np.pi)-np.pi/2   #  map angles to between +/- pi if the rod flips

    alpha_zeros = np.where(np.diff(np.sign(alpha)))[0] # find the indicies in the alpha list where alpha crosses sign (i.e. goes through zero).
    beta_before_zero=np.array([beta[i] for i in alpha_zeros]) # find the list of betas just before each alpha=0
    beta_after_zero=np.array([beta[i] for i in alpha_zeros+1]) # find the list of betas just after each alpha=0
    dot_beta_at_zero=(beta_after_zero-beta_before_zero)/dt # approximate beta_dot at each alpha=0

    plt.xlim([-np.pi,np.pi])
    plt.ylim([-50,50])
    plt.xlabel('$\\beta$ (rad)')
    plt.ylabel('$\\dot{\\beta} (rad/s)$')
    plt.scatter((beta_before_zero+beta_after_zero)/2,dot_beta_at_zero) #Plot poincare section: i.e. scatter plot of beta and beta_dot where, at each point, alpha=0.#
    plt.show()
```
